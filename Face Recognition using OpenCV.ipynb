```python
import cv2
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Assuming the file is in the same directory as the notebook
image = cv2.imread('C:\\Users\\siddh_ch0ux8e\\OneDrive\\MAIN\\OneDrive\\Documents\\PROJECT\\PRJ Face Detection\\Data Set\\people1.jpg')

# Check if the image was loaded correctly
if image is None:
    print("Error: Could not load the image.")
else:
    # Display the loaded image using matplotlib (Jupyter-friendly)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Original Image")
    plt.show()
    
    # Resize the image for consistent processing
    image = cv2.resize(image, (800, 600))
    
    # Convert the image to grayscale
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Load the Haar Cascade classifier
    # Full path to the file
    face_detector = cv2.CascadeClassifier('C:\\Users\\siddh_ch0ux8e\\OneDrive\\MAIN\\OneDrive\\Documents\\PROJECT\\PRJ Face Detection\\Cascades\\haarcascade_frontalface_default.xml')
    
    # Check if the classifier was loaded correctly
    if face_detector.empty():
        print("Error: Could not load the Haar Cascade classifier. Make sure the file is in the correct path.")
    else:
        # Perform face detection
        # The detectMultiScale function returns a list of (x, y, width, height) coordinates for each detected face.
        detections = face_detector.detectMultiScale(image_gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
        
        print(f"Number of faces detected: {len(detections)}")
        
        # Draw bounding boxes around the detected faces
        for (x, y, w, h) in detections:
            cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
            
        # Display the image with the bounding boxes using cv2.imshow
        cv2.imshow('Face Detection', image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
```
